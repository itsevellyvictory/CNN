```python
from tensorflow.keras.utils import to_categorical
```

```python
y_cat_train = to_categorical(y_train)
```

```python
y_cat_test = to_categorical(y_test)
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
x_train.shape
```

```python
x_train = x_train.reshape(60000,28,28,1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_test = x_test.reshape(10000,28,28,1)
```

```python
x_test.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout
```

```python
model = Sequential()

model.add(Conv2D(filters=32,kernel_size=(3,3),activation = 'linear',input_shape=(28,28,1),padding='same'))
model.add(tf.keras.layers.LeakyReLU(alpha = 0.1))
model.add(MaxPool2D(pool_size=(2,2),padding='same'))
model.add(Conv2D(64,(3,3),activation = 'linear',padding = 'same'))
model.add(tf.keras.layers.LeakyReLU(alpha = 0.1))
model.add(MaxPool2D(pool_size=(2,2),padding='same'))
model.add(Conv2D(128,(3,3),activation = 'linear',padding = 'same'))
model.add(tf.keras.layers.LeakyReLU(alpha = 0.1))
model.add(MaxPool2D(pool_size=(2,2),padding='same'))
model.add(Flatten())
model.add(Dense(128,activation = 'relu'))
model.add(tf.keras.layers.LeakyReLU(alpha = 0.1))
model.add(Dense(10,activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer= 'adam', metrics = ['accuracy'])
```

```python
model.fit(x_train, y_cat_train, batch_size= 64, epochs = 30, validation_data=(x_test, y_cat_test))
```

```python
model.metrics_names
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses
```

```python
losses[['loss','val_loss']].plot()
```

```python
losses[['accuracy','val_accuracy']].plot()
```

```python
['loss', 'accuracy']

```

```python
model.evaluate(x_test,y_cat_test)
```

```python
# Garantir que x_test está pré-processado (float / 255, com canal)
print("Shape x_test:", x_test.shape)
print("Shape y_test:", y_test.shape)

# Prever probabilidades para cada classe
y_test_pred = model.predict(x_test)  # shape (10000, 10)

# Converter para rótulos previstos (classe com maior probabilidade)
y_test_pred_classes = np.argmax(y_test_pred, axis=1)  # shape (10000,)

print("Shape y_test_pred_classes:", y_test_pred_classes.shape)
print("Valores únicos em y_test_pred_classes:", np.unique(y_test_pred_classes))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("\nRelatório de classificação no TESTE:")
print(classification_report(y_test, y_test_pred_classes))

print("\nMatriz de confusão:")
print(confusion_matrix(y_test, y_test_pred_classes))
```
